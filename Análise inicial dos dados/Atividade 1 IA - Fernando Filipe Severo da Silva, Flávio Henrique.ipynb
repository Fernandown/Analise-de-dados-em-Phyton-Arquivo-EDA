```python
# Import das bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

#carregar os dados básicos e os dados dentro do pandasDataFrame
dt = load_breast_cancer() #coloca os dados da função dentro da variável dt
df = pd.DataFrame(dt.data, columns=dt.feature_names) #colocando os dados no dataframe
df['target'] = dt.target 

#transformar dados para o formato de tabela
print(df) #mostra os dados que estão no data frame df
```

```python
#Heatmap
plt.figure(figsize=(20, 12)) #tamanho da tabela do matplotlib
sns.heatmap(df.corr(), annot=True) #recebe o dataframe de sklearn e reproduz os dados
```

```python
#histograma
df.hist(figsize=(30, 15)) #tamanho da tabela do matplotlib
plt.show() #mostra em formato historigrama - dados já inseridos em plt
```

```python
#BloxPlot
plt.figure(figsize=(12, 8)) #tamanho da tabela do matplotlib
sns.boxplot(data=df.drop('target', axis=1)) #colocando o dataframe limitado na variável data para se adequar a tabela
plt.xticks(rotation=90)
plt.show()
```

```python
#ViolinPlot
plt.figure(figsize=(12, 8)) #tamanho da tabela
sns.violinplot(data=df.drop('target', axis=1)) #colocando o dataframe com condições na função de criação da tabela
plt.xticks(rotation=90) #nomes que ficam em baixo da tabela, muda a orientação deles
plt.show() #manda mostrar a tabela
```

```python
#ScatterPlot
plt.figure(figsize=(12, 8)) #define o tamanho do gráfico
sns.scatterplot(x='mean radius', y='mean texture', hue='target', data=df) #coloca os dados do df, nomeia o lado X e Y
plt.show() #mostra a tabela
```
